```python
main_df_info = pd.read_csv("Students_info_0.csv")
main_df_marks = pd.read_csv("Students_marks_0.csv")
for i in range(1, 10):
    df_info = pd.read_csv(f"Students_info_{i}.csv") 
    df_marks = pd.read_csv(f"Students_marks_{i}.csv")
    main_df_info = pd.concat([main_df_info, df_info])
    main_df_marks = pd.concat([main_df_marks, df_marks])
df = pd.concat([main_df_info, main_df_marks], axis=1)    
df.head()
```

```python
def GetGrade(average_mark):
    if average_mark >= 90:
        return "A"
    if average_mark >= 80:
        return "B"
    if average_mark >= 70:
        return "C"
    if average_mark >= 60:
        return "D"
    if average_mark >= 50:
        return "E"
    return "F"


df["avg_mark"] = df.apply(lambda x: (x["math score"] + x["reading score"] + x["writing score"]) / 3, axis=1)
df["Grade"] = df["avg_mark"].apply(GetGrade)
df = df.drop("avg_mark", axis=1)
df.head()
```
